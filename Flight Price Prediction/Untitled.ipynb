```python
final_df.insert(3,'date',final_df.Date_of_Journey.str.split('/').str[0])
final_df.insert(4,'Month',final_df.Date_of_Journey.str.split('/').str[1])
final_df.insert(5,'Year',final_df.Date_of_Journey.str.split('/').str[2])
final_df.head()
```

```python
final_df.Date = final_df.date.astype(int)
final_df.Month = final_df.Month.astype(int)
final_df.Year = final_df.Year.astype(int)
final_df.info()
```

```python
final_df.drop(['Date_of_Journey'], axis = 1, inplace=True)
final_df.head()
```

```python
backup = final_df
```

```python
final_df.Arrival_Time = final_df.Arrival_Time.apply(lambda x:x.split(' ')[0])
final_df.head()
```

```python
final_df.insert(9,'Arrival_hr', final_df.Arrival_Time.apply(lambda x:x.split(':')[0]))
final_df.insert(10,'Arrival_min', final_df.Arrival_Time.apply(lambda x:x.split(':')[1]))
final_df.Arrival_min = final_df.Arrival_min.astype(int)
final_df.Arrival_hr = final_df.Arrival_hr.astype(int)
final_df.drop(['Arrival_Time'], axis = 1, inplace=True)
final_df.head()
```

```python
backup = final_df
```

```python
final_df.insert(11,'Dep_hr', final_df.Dep_Time.apply(lambda x:x.split(':')[0]))
final_df.insert(12,'Dep_min', final_df.Dep_Time.apply(lambda x:x.split(':')[1]))
final_df.Dep_min = final_df.Dep_min.astype(int)
final_df.Dep_hr = final_df.Dep_hr.astype(int)
final_df.drop(['Dep_Time'], axis = 1, inplace=True)
final_df.head()
```
